# <span style="font-family:cursive; color:#0F2557;"><b>6. Similarity Calculation</b></span>

**Nominal Attributes:**
<br>
<ul>
<li><b>workclass:</b> Categorical data representing employment status, with no inherent order.</li>
<li><b>marital_status:</b> Categorical data representing marital status, with no inherent order.</li>
<li><b>occupation:</b>  Categorical data representing different job types, with no inherent order.</li>
<li><b>relationship:</b>  Categorical data representing the type of family relationship, with no inherent order.</li>
<li><b>race:</b>  Categorical data representing different races, with no inherent order.</li>
<li><b>sex:</b> Categorical data representing gender, with no inherent order.</li>
<li><b>native_country:</b> Categorical data representing the country of origin, with no inherent order.</li>
<li><b>high_income:</b>  Categorical data representing whether income is high or low, with no inherent order (binary classification).</li> 
</ul>
<b>Ordinal Attributes: </b>
<br>
<ul>
<li><b>education:</b>  Categorical data representing different levels of education, which have a meaningful order but not necessarily equal intervals.</li>
</ul>   
<b>Ratio Attributes:</b>
<br>
<ul>
<li><b>age:</b>  Numerical data representing age, with a meaningful zero point and allowing for the comparison of ratios.</li>
<li><b>fnlwgt:</b>  Numerical data representing the final weight, used by the census for sampling purposes.</li>
<li><b>education_num:</b>  Numerical representation of education level, with meaningful intervals and a meaningful zero point.</li>
<li><b>capital_gain:</b>  Numerical data representing capital gains, with a meaningful zero point.</li>
<li><b>capital_loss:</b>  Numerical data representing capital losses, with a meaningful zero point.</li>
<li><b>hours_per_week:</b>  Numerical data representing the number of hours worked per week, with a meaningful zero point.</li>
</ul>

```python
# Identify attribute types
ordinal_attributes = ['education']
nominal_attributes = ['workclass', 'marital_status', 'occupation', 'relationship', 'race', 'sex', 'native_country', 'high_income']
ratio_attributes = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']
```

```python
# Similarity functions
def nominal_similarity(p, q):
    return 1 if p == q else 0

def ordinal_similarity(p, q, n):
    d = abs(p - q) / (n - 1)
    s = 1 - d
    return s

def ratio_similarity(p, q):
    d = abs(p - q)
    s = 1 / (1 + d)
    return s

def calculate_similarity_matrix(df, attribute, similarity_func, is_ordinal=False):
    values = sampled_df[attribute].values
    unique_values = sorted(sampled_df[attribute].unique()) if is_ordinal else None
    n = len(unique_values) if unique_values is not None else None

    similarity_matrix = np.zeros((len(values), len(values)))

    for i in range(len(values)):
        for j in range(len(values)):
            if is_ordinal:
                similarity_matrix[i, j] = similarity_func(values[i], values[j], n)
            else:
                similarity_matrix[i, j] = similarity_func(values[i], values[j])

    return similarity_matrix
```

```python
# Calculate similarity matrices for each attribute type
similarity_matrices = []

# Nominal attributes
for attribute in nominal_attributes:
    similarity_matrices.append(calculate_similarity_matrix(sampled_df, attribute, nominal_similarity))

# Ordinal attributes
for attribute in ordinal_attributes:
    similarity_matrices.append(calculate_similarity_matrix(sampled_df, attribute, ordinal_similarity, is_ordinal=True))

# Ratio attributes
for attribute in ratio_attributes:
    similarity_matrices.append(calculate_similarity_matrix(sampled_df, attribute, ratio_similarity))

# Combine similarity matrices into a single matrix
combined_similarity_matrix = np.zeros((len(sampled_df), len(sampled_df)))

# Weighted sum of all similarity matrices (weights can be adjusted as needed)
for matrix in similarity_matrices:
    combined_similarity_matrix += matrix

# Normalize combined similarity matrix
combined_similarity_matrix /= len(similarity_matrices)
np.set_printoptions(precision=2)
print("Combined Similarity Matrix:\n", combined_similarity_matrix)
```

```python
#Creating Dataframe for the final values
final_matrix = pd.DataFrame(combined_similarity_matrix)
final_matrix
```

```python
#Writing the combined similarity to output file
final_matrix.to_csv('Adult_dataset_Similarity_Matrix.csv')
```

```python
# Store actual distance matrix to file output/distance_output_actual.txt
distance_matrix = 1 - combined_similarity_matrix
np.savetxt('adult_dataset_distance_output_actual.txt', distance_matrix, fmt='%.3f')

# Convert combined distance matrix to binary matrix, threshold 0.5
binary_distance_matrix = (distance_matrix >= 0.5).astype(int)

# Store binary distance matrix to file output/distance_output_binary.txt
np.savetxt('adult_dataset_distance_output_binary.txt', binary_distance_matrix, fmt='%d')
```

```python
# Store actual similarity matrix to file output/similarity_output_actual.txt
np.savetxt('adult_dataset_similarity_output_actual.txt', combined_similarity_matrix, fmt='%.3f')

# Convert combined similarity matrix to binary matrix, threshold 0.5
binary_similarity_matrix = (combined_similarity_matrix >= 0.5).astype(int)

# Store binary similarity matrix to file output/similarity_output_binary.txt
np.savetxt('adult_dataset_similarity_output_binary.txt', binary_similarity_matrix, fmt='%d')

print("Process completed successfully.")
```
